```python
import numpy as np
import pandas as pd

# 1. Definição dos limites aproximados do Sudeste (Lat/Lon)
# Minas, SP, RJ e ES ficam aproximadamente entre:
LAT_LIMITS = (-25.0, -14.0) 
LON_LIMITS = (-53.0, -39.0)

N_HOSPITAIS = 100

# 2. Gerar coordenadas aleatórias dentro desses limites
np.random.seed(42)
lats = np.random.uniform(LAT_LIMITS[0], LAT_LIMITS[1], N_HOSPITAIS)
lons = np.random.uniform(LON_LIMITS[0], LON_LIMITS[1], N_HOSPITAIS)

# 3. Criar o DataFrame com Prioridades e Pesos (Requisitos do Projeto 2)
dados_hospitais = pd.DataFrame({
    'id': range(N_HOSPITAIS),
    'lat': lats,
    'lon': lons,
    # 2 = Medicamento Crítico (Urgente), 1 = Insumo Regular
    'prioridade': np.random.choice([1, 2], N_HOSPITAIS, p=[0.7, 0.3]),
    'carga_kg': np.random.randint(2, 15, N_HOSPITAIS)
})

# O primeiro ponto (índice 0) será o nosso "Hospital Central / Centro de Distribuição"
dados_hospitais.loc[0, 'prioridade'] = 0 # Marcar como Ponto de Partida
print(f"{N_HOSPITAIS} pontos gerados no mapa do Sudeste.")
```

```python
# Configurações da Janela
LARGURA, ALTURA = 800, 600
MARGEM = 60

def geo_para_pixel(lat, lon):
    # Escalonamento linear baseado nos limites que definimos
    x = int(((lon - LON_LIMITS[0]) / (LON_LIMITS[1] - LON_LIMITS[0])) * (LARGURA - 2 * MARGEM) + MARGEM)
    y = int(((LAT_LIMITS[1] - lat) / (LAT_LIMITS[1] - LAT_LIMITS[0])) * (ALTURA - 2 * MARGEM) + MARGEM)
    return (x, y)

# Converter todos os 100 hospitais para pixels
pontos_pixels = [geo_para_pixel(row['lat'], row['lon']) for _, row in dados_hospitais.iterrows()]
```

```python
import pygame

pygame.init()
tela = pygame.display.set_mode((LARGURA, ALTURA))
pygame.display.set_caption("Logística Hospitalar - 100 Unidades Sudeste")

# Loop de Visualização
rodando = True
while rodando:
    for evento in pygame.event.get():
        if evento.type == pygame.QUIT: rodando = False

    tela.fill((255, 255, 255))

    # Desenhar os 100 pontos
    for i, pos in enumerate(pontos_pixels):
        cor = (100, 100, 100) # Padrão: Cinza
        tamanho = 4
        
        if i == 0: # Hospital Central
            cor = (0, 200, 0)
            tamanho = 8
        elif dados_hospitais.iloc[i]['prioridade'] == 2: # Crítico
            cor = (255, 0, 0)
            tamanho = 5

        pygame.draw.circle(tela, cor, pos, tamanho)

    pygame.display.flip()

pygame.quit()
```
