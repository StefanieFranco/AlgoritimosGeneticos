```python
import pandas as pd
import numpy as np
import os

# 1. Garantir que a pasta existe
os.makedirs('data', exist_ok=True)

# 2. Configurações
N_DESTINOS = 100
np.random.seed(42)

# 3. Gerando os dados
dados = {
    'id': range(N_DESTINOS),
    'x': np.random.randint(50, 750, N_DESTINOS),
    'y': np.random.randint(50, 550, N_DESTINOS),
    'prioridade': np.random.choice([1, 2], N_DESTINOS, p=[0.7, 0.3]),
    'demanda_kg': np.random.randint(2, 12, N_DESTINOS),
    # Nova coluna: Tipo de Destino (70% Hospitais, 30% Atendimento Domiciliar/Casa)
    'tipo': np.random.choice(['hospital', 'casa'], N_DESTINOS, p=[0.7, 0.3]),
    # Nova coluna: Janela de Atendimento
    'horario': np.random.choice(['24h', 'comercial'], N_DESTINOS, p=[0.6, 0.4])
}

df_destinos = pd.DataFrame(dados)

# Ajuste do Ponto Zero (Centro de Distribuição)
df_destinos.loc[0, 'tipo'] = 'central'
df_destinos.loc[0, 'horario'] = '24h'
df_destinos.loc[0, 'prioridade'] = 0
df_destinos.loc[0, 'demanda_kg'] = 0

# 4. Salvar como destinos.csv
df_destinos.to_csv('data/destinos.csv', index=False)
print("Arquivo data/destinos.csv gerado com sucesso!")
df_destinos.head()
```

```python
import pandas as pd

# Definindo a frota conforme sua ideia
frota = []

# 5 Motos (Agilidade e baixo custo, pouca carga)
for i in range(5):
    frota.append({'id': f'MOTO_{i+1}', 'tipo': 'moto', 'autonomia_km': 150, 'capacidade_kg': 20, 'custo_km': 1.0})

# 5 Mini Caminhões (Equilíbrio)
for i in range(5):
    frota.append({'id': f'MINI_{i+1}', 'tipo': 'mini_caminhao', 'autonomia_km': 400, 'capacidade_kg': 200, 'custo_km': 2.5})

# 5 Caminhões (Alta carga, custo elevado)
for i in range(5):
    frota.append({'id': f'TRUCK_{i+1}', 'tipo': 'caminhao', 'autonomia_km': 800, 'capacidade_kg': 1000, 'custo_km': 5.0})

df_veiculos = pd.DataFrame(frota)
df_veiculos.to_csv('data/vehicles.csv', index=False)
print("Frota de 15 veículos gerada em data/vehicles.csv")
```

```python
import pandas as pd
import numpy as np
import os

# 1. Configurações
N_ENTREGAS = 99 # 100 destinos menos a base (ID 0)
np.random.seed(42)

entregas = []

for i in range(1, 100): # Começa do 1 para pular o Centro de Distribuição
    sorteio_carga = np.random.random()
    
    # Lógica para ocupar a frota variada:
    if sorteio_carga < 0.5:
        # 50% são Leves (Ideal para MOTOS)
        peso = np.random.uniform(0.1, 5.0)
        descricao = "Medicamentos/Vacinas"
    elif sorteio_carga < 0.85:
        # 35% são Médias (Ideal para MINI CAMINHÕES)
        peso = np.random.uniform(15.0, 60.0)
        descricao = "Insumos/Soro/Gaze"
    else:
        # 15% são Pesadas (Exige CAMINHÕES)
        peso = np.random.uniform(150.0, 500.0)
        descricao = "Equipamentos Hospitalares"

    entregas.append({
        'id': f'DELIV_{i:03d}',
        'destino_id': i,
        'peso_kg': round(peso, 2),
        'descricao': descricao,
        'valor_declarado': round(np.random.uniform(50, 5000), 2) # Útil para seguros/prioridade
    })

# 2. Criar DataFrame e Salvar
df_deliveries = pd.DataFrame(entregas)
df_deliveries.to_csv('data/deliveries.csv', index=False)

print("Arquivo data/deliveries.csv gerado com sucesso!")
print(df_deliveries.head())
```
